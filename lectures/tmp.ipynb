```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_theme()

plt.close('all')


N=1000

"""1 code"""
x = np.random.randn(N)
y = x + 1 +np.sqrt(3)*np.random.randn(N)
data1 = pd.DataFrame(data={'x':x, 'y':y})
sns.jointplot(data=data1, x="x", y="y", color='blue')


"""2 code"""
y = 1 + 2*np.random.randn(N)
x = (y-1)/4 + np.sqrt(3) * np.random.rand(N)/2
data2 = pd.DataFrame(data={'x':x, 'y':y})
sns.jointplot(data=data2, x="x", y="y", color='red')


"""3 code"""
z = np.random.randn(N)
y = z + 1 + np.sqrt(3)*np.random.randn(N)
x = z
data3 = pd.DataFrame(data={'x':x, 'y':y})
sns.jointplot(data=data3, x="x", y="y", color='green')
```
